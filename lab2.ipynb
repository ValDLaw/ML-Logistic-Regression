```python
from PIL import Image
import pandas as pd
tuple_list = []
for _ in range(1,9):
    for impath in os.listdir(f'./output/test/{_}/'):
        imp = os.path.join(f'./output/test/{_}/', impath)
        im = Image.open(imp)
        mat = np.zeros((28, 157), dtype=np.uint8)
        w, h = im.size
        # print(mat[:,-w:].shape == np.array(im).shape)
        mat[:,-w:] = np.array(im)
        print(mat.shape)
        im = Image.fromarray(mat)
        im.save(imp)
```

```python
import pandas as pd

tuple_list = []     # (feature_vector, class_label)
datapath = "output/test"
for file in os.listdir(datapath):
    class_label = file
    n = os.path.join(datapath, file)
    if os.path.isdir(n):
        for img in os.listdir(n):
            img_path = os.path.join(n, img)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            
            img = cv2.resize(img, (28*2, 157*2))
            
            # Aplicar algún preprocesamiento si es necesario (puede ser necesario según tus imágenes)
            # Por ejemplo, puedes aplicar redimensionamiento, filtrado, etc.
            
            # Extraer características (en este caso, HOG)
            fd, hog_image = hog(img, orientations=8, pixels_per_cell=(16, 16),
                                cells_per_block=(1, 1), visualize=True)
            fd = fd.astype('float32')
            tuple_list.append((fd, class_label))

df = pd.DataFrame.from_records(tuple_list, columns=['features', 'labels'])
display(df)
```

```python
X = np.array([v for v in df['features']])#np.array(features)
y = np.array(df['labels'].values)#np.array(labels)

# df_feat = pd.DataFrame(X, columns=list(range(len(X[0]))))
# df_feat['class'] = y
# tuple_list = []
# for col in df_feat.columns:
#     tuple_list.append((col, df_feat[col].var()))
#     # print(f'{col} -> {df_feat[col].var()}')

# var_feat = sorted(tuple_list, key=lambda t: t[1])
```

```python
from sklearn.model_selection import train_test_split
SEED = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, shuffle=True, test_size=0.2, random_state=SEED)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, stratify=y_test, shuffle=True, test_size=0.5, random_state=SEED)

print(X_train.shape, X_test.shape, X_val.shape)
```

```python
param_grid = {
    'penalty' : ['l1', 'l2', 'elasticnet'],
    'C' : [5, 7.5, 10, 12.5, 15], # 0.1, 1, 100
    'multi_class' : ['ovr', 'multinomial'],
    'max_iter' : [1000, 2000],
    'solver': ['saga']
}

lr = LogisticRegression()
clf = GridSearchCV(lr, param_grid, cv=5)
clf.fit(X_train, y_train)
best_params = clf.best_params_
print(best_params)
```

```python
best_model = clf.best_estimator_
best_model
```

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn.model_selection import KFold
kf = KFold(n_splits=5, random_state=SEED, shuffle=True)
for i, (train_index, test_index) in enumerate(kf.split(X)):
   print(f'Fold {i}')
   model = LogisticRegression(**clf.best_params_)
   X_train, y_train = X[train_index], y[train_index]

   XX, yy = X[test_index], y[test_index]
   X_test, X_val, y_test, y_val = train_test_split(XX, yy, stratify=yy, test_size=0.5, shuffle=True, random_state=SEED)
   model.fit(X_train, y_train)

   print(classification_report(y_train, model.predict(X_train), labels=['1','2','3','4','5','6','7','8']))
   print(classification_report(y_test, model.predict(X_test), labels=['1','2','3','4','5','6','7','8'])) 
   print(classification_report(y_val, model.predict(X_val), labels=['1','2','3','4','5','6','7','8']))
```

```python
def report(model, X, y):
    y_train_pred = best_model.predict(X)
    cm_train = confusion_matrix(y, y_train_pred)
    classes = ['Un dígito', 'Dos dígitos', 'Tres dígitos', 'Duatro dígitos', 'Cinco dígitos', 'Seis dígitos', 'Siete dígitos', 'Ocho dígitos', 'Nueve dígitos'] # Continuar con el número apropiado de clases.

    print("Confusion Matrix - Training Set:")
    print(np.array2string(cm_train, separator=', ', formatter={'int':lambda x: f'{x:5}'}))
    print("Classification Report")
    print(classification_report(y, y_train_pred, labels=[1,2,3,4,5,6,7,8]))
```
